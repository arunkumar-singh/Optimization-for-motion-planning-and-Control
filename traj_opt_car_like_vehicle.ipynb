# Trajectory Planning for a Car-Like Robot Using Gauss Newton Method



# Section 1: Trajectory Paramterization

A car-like robot moves by commanding forward ($v_t$) and angular velocity $\omega_t$ at any time $t$. Suppose, the car is at position $(x_t, y_t)$ with orientation $\psi_t$. Then if we apply the forward and angular velocity for $\Delta t$ time, the new position of the car will be

\begin{align}
    x[k+] = x[k]+v[k]\cos\theta[k]\Delta t, y[k+] = y[k]+v[k]\sin\theta[k]\Delta t \\
    \theta[k+1] = \theta[k]+\omega[k]\Delta t
\end{align}

Now, imagine, we are given a sequence of linear and angular velocities, $(v[0], v[1], \dots v[n])$, $(\omega[0], \omega[1], \dots \omega[n])$, then we can express any position of the robot at time step $n$ in the following manner.

\begin{align}
    x[n] = x[0]+\sum_{k=1}^{k = n-1} v\cos(\theta[0]+\sum_{k=1}^{k=n-1}\omega[k]\Delta t )\Delta t)\\
    y[n] = y[0]+\sum_{k=1}^{k = n-1} v\sin(\theta[0]+\sum_{k=1}^{k=n-1}\omega[k]\Delta t )\Delta t)\\
    \theta[n] = \theta[0]+\sum_{k=1}^{k=n-1}\omega[k]\Delta t
\end{align}



Let's simulate the above equations



```python
import jax.numpy as jnp
import numpy as np
import matplotlib.pyplot as plt


x_0 = 0.0
y_0 = 0.0
theta_0 = 0.0

v = 0.3*jnp.ones(5)
omega = 0.2*jnp.ones(5)
delta_t = 0.2

def compute_car_traj(x_0, y_0, theta_0, v, omega, delta_t   ):

  theta_t = theta_0+jnp.cumsum(omega*delta_t)
  x_t = x_0+jnp.cumsum(v*jnp.cos(theta_t)*delta_t   )
  y_t = y_0+jnp.cumsum(v*jnp.sin(theta_t)*delta_t   )

  return x_t, y_t


x_t, y_t = compute_car_traj(x_0, y_0, theta_0, v, omega, delta_t   )


plt.figure(1)
plt.plot(x_t, y_t, '-om', linewidth = 3.0)
plt.axis('equal')
plt.show()


```
